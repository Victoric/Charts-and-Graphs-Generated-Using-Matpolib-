```python
import matplotlib.pyplot as plt

# Data
data = {
    'Gender': {
        'Male': 16,
        'Female': 14,
    },
    'Age': {
        '18-24 years': 18,
        '25-34 years': 10,
        '35-44 years': 2,
    },
    'Educational Background': {
        'Humanities/Social Sciences': 12,
        'Sciences/Engineering': 10,
        'Business/Economics': 8,
        'Other': 0,
    }
}

# Function to plot pie chart
def plot_pie_chart(data, title):
    labels = list(data.keys())
    sizes = list(data.values())
    percentages = [round((size / sum(sizes)) * 100, 1) for size in sizes]
    legend_labels = [f"{label} ({percentage} %)" for label, percentage in zip(labels, percentages)]

    fig, ax = plt.subplots()
    ax.pie(sizes, labels=labels, autopct='%.1f%%', startangle=90)
    ax.axis('equal')
    ax.set_title(title)
    ax.legend(legend_labels, loc="upper right")

# Plotting
for title, data_dict in data.items():
    plot_pie_chart(data_dict, title)

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Data
data = {
    'Participant ID': list(range(1, 31)),
    'Age': [30, 22, 20, 23, 21, 24, 27, 22, 32, 23, 25, 24, 20, 30, 29, 23, 21, 25, 37, 22, 19, 23, 21, 25, 20, 22, 19, 25, 35, 21],
    'Gender': ['M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'M'],
    'Pre-test Score': [60, 62, 58, 64, 61, 63, 59, 65, 60, 62, 58, 64, 61, 63, 59, 65, 60, 62, 58, 64, 61, 63, 59, 65, 60, 62, 58, 64, 61, 61],
    'Post-test Score': [75, 78, 73, 80, 77, 79, 74, 81, 75, 78, 73, 80, 76, 79, 74, 81, 75, 78, 73, 80, 76, 79, 74, 81, 75, 78, 73, 80, 76, 76],
    'Improvement': [15, 16, 15, 16, 16, 16, 15, 16, 15, 16, 15, 16, 15, 16, 15, 16, 15, 16, 15, 16, 15, 16, 15, 16, 15, 16, 15, 16, 15, 15],
}

df = pd.DataFrame(data)

# Function to plot bar chart
def plot_bar_chart(x, y, xlabel, ylabel, title):
    fig, ax = plt.subplots()
    ax.bar(x, y)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title)

# Plotting improvement by participant ID
plot_bar_chart(df['Participant ID'], df['Improvement'], 'Participant ID', 'Improvement', 'Improvement by Participant ID')

# Plotting pre-test and post-test scores by participant ID
fig, ax = plt.subplots()
ax.bar(df['Participant ID'] - 0.2, df['Pre-test Score'], width=0.4, label='Pre-test Score')
ax.bar(df['Participant ID'] + 0.2, df['Post-test Score'], width=0.4, label='Post-test Score')
ax.set_xlabel('Participant ID')
ax.set_ylabel('Score')
ax.set_title('Pre-test and Post-test Scores by Participant ID')
ax.legend()

plt.show()
```
